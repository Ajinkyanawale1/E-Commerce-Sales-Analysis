```python
#_________________________Data Visualization__________________________________
```

```python
df['Month_Name'] = df['Date'].dt.strftime('%B')
```

```python
sales_over_time = df.resample('M', on='Date')['Amount'].sum().reset_index()
plt.figure(figsize=(10,6))
sns.lineplot(x='Date', y='Amount', data=sales_over_time)
plt.title('Total Sales Over Time (Monthly Aggregation)')
plt.xlabel('Date')
plt.ylabel('Total Sales (Amount)')
plt.xticks(rotation=45)
plt.show()
```
